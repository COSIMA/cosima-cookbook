# Annually Averaged Scalar Timeseries

This notebook shows how we extract information from ocean_scalar to plot timeseries. The data are resampled onto annual averages.

**Firstly,** load in the required libraries:

````python
%matplotlib inline

import cosima_cookbook as cc
import matplotlib.pyplot as plt
````

**Next,** we show to plot a single variable from a single experiment. The variable is loaded using ```get_nc_variable```. Note that we reference dates to 1700, so that we can use pandas to ```resample``` the data onto annual frequency. 

````python
expt = '1deg_jra55v13_ryf9091_spinup_A'
variable = 'temp_global_ave'
darray = cc.get_nc_variable(expt,
                             'ocean_scalar.nc',
                             variable,
                             time_units='days since 1700-01-01')
annual_average = darray.resample('A', 'time').compute()
````

**Then,** the data can be plotted as you see fit:

````python
plt.figure(figsize=(8,5))
annual_average.plot()
plt.title(expt)
````
